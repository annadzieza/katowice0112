```
import pandas as pd
import datetime
import pandas_datareader.data as web
from pandas import Series, DataFrame

#Ładowanie danych z bazy notowań na yahoo
#Ceny akcji IBM w latach 2010-2018

start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2018, 1, 12)

df = web.DataReader("IBM", 'yahoo', start, end)
df.head()
```

```
list(df.columns)
```

```
#Cena zamknięcia akcji oraz średnia ważona
close_px = df['Adj Close']
mavg = close_px.rolling(window=100).mean()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style

import matplotlib as mpl
mpl.rc('figure', figsize=(8, 7))
mpl.__version__

style.use('ggplot')

close_px.plot(label='IBM')
mavg.plot(label='średnia ważona')
plt.legend()
```

```
#Stopy zwrotu z akcji IBM - obliczenie i wyświetlenie na wykresie
rets = close_px / close_px.shift(1) - 1
rets.plot(label='return')
```

```
#Dane z notowań innych firm technologicznych do porównania
dfcomp = web.DataReader(['IBM', 'GE', 'GOOG', 'AAPL', 'MSFT'],'yahoo',start=start,end=end)['Adj Close']
```

```
#Sprawdzenie korelacji między procentowymi zmianami cen akcji poszczególnych firm technologicznych - sprawdzenie czy ceny akcji firm zależą od siebie
retscomp = dfcomp.pct_change()

corr = retscomp.corr()
```

```
#Wykres korelacji procentowych zmian cen IBM i APPLE
plt.scatter(retscomp.IBM, retscomp.AAPL)
plt.xlabel('Returns IBM')
plt.ylabel('Returns APPLE')
```

```
#Wykres korelacji cen akcji poszczególnych firm technologicznych
pd.plotting.scatter_matrix(retscomp, diagonal='kde', figsize=(10, 10));
```

```
#Heatmapa do lepszej wizualizacji korelacji
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
```

```
#Wykres zależności stóp zwrotu z akcji i ryzyka inwestycji (mierzonego odchyleniem standardowym ceny akcji)
plt.scatter(retscomp.mean(), retscomp.std())
plt.xlabel('Expected returns')
plt.ylabel('Risk')
for label, x, y in zip(retscomp.columns, retscomp.mean(), retscomp.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```

```
dfreg = df.loc[:,['Adj Close', 'Volume']]
len(dfreg)
```

```
import math
import numpy as np
from sklearn import preprocessing

#Uzupełnianie brakujących danych
dfreg.fillna(value=-99999, inplace=True)
#Wybór 1% danych z bazy do prognozy
forecast_out = int(math.ceil(0.01 * len(dfreg)))
#Labels - prognozujemy wartość kolumny 'Adj Close', czyli cenę rynkową akcji
forecast_col = 'Adj Close'
dfreg['label'] = dfreg[forecast_col].shift(-forecast_out)
X = np.array(dfreg.drop(['label'], 1))
#Skalowanie danych
X = preprocessing.scale(X)
#Zdefiniowanie kolumn
X = X[:-forecast_out]
y = np.array(dfreg['label'])
y = y[:-forecast_out]
```

```
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

from sklearn.ensemble import RandomForestRegressor
```

```
from sklearn.model_selection import train_test_split
#Podział zbioru na dane testowe i trenowane
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
#Regresja liniowa - dopasowanie danych
clfreg = LinearRegression(n_jobs=-1)
clfreg.fit(X_train, y_train)
#Regresja brzegowa - dopasowanie danych
clfpoly2 = make_pipeline(PolynomialFeatures(2), Ridge())
clfpoly2.fit(X_train, y_train)
#Las Losowy - dopasowanie danych
random_forest = RandomForestRegressor()
random_forest.fit(X_train, y_train)
#Regresja KNN
clfknn = KNeighborsRegressor(n_neighbors=2)
clfknn.fit(X_train, y_train)
```

```
confidencereg = clfreg.score(X_test, y_test)
confidencepoly2 = clfpoly2.score(X_test,y_test)
confidencerandom_forest = random_forest.score(X_test,y_test)
confidenceknn = clfknn.score(X_test, y_test)

print('The linear regression confidence is ', confidencereg)
print('The quadratic regression 2 confidence is ', confidencepoly2)
print('The random forest regressor confidence is ', confidencepoly3)
print('The knn regression confidence is ', confidenceknn)
```
